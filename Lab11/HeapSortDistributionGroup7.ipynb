# Experimental analysis of the Heap Sort execution time 

```
#tomado de: https://www.geeksforgeeks.org/python-program-for-heap-sort/

def heapify(arr, n, i):
    largest = i  # Initialize largest as root
    l = 2 * i + 1     # left = 2*i + 1
    r = 2 * i + 2     # right = 2*i + 2
  
    # See if left child of root exists and is
    # greater than root
    if l < n and arr[i] < arr[l]:
        largest = l
  
    # See if right child of root exists and is
    # greater than root
    if r < n and arr[largest] < arr[r]:
        largest = r
  
    # Change root, if needed
    if largest != i:
        arr[i],arr[largest] = arr[largest],arr[i]  # swap
  
        # Heapify the root.
        heapify(arr, n, largest)
  
# The main function to sort an array of given size
def heapSort(arr):
    n = len(arr)
  
    # Build a maxheap.
    # Since last parent will be at ((n//2)-1) we can start at that location.
    for i in range(n // 2 - 1, -1, -1):
        heapify(arr, n, i)
  
    # One by one extract elements
    for i in range(n-1, 0, -1):
        arr[i], arr[0] = arr[0], arr[i]   # swap
        heapify(arr, i, 0)
        
    return(arr)
```

```
A = [1,2,3,4,5,6,7,8]
heapSort(A)
print(A)

A = [6,7,2,5,4,3,8,1]
heapSort(A)
print(A)

A = [8,7,6,5,4,3,2,1]
heapSort(A)
print(A)
```

```
#tomado de: https://www.geeksforgeeks.org/python-program-for-heap-sort/
def heapify_steps(arr, n, i,steps):
    largest = i 
    l = 2 * i + 1 
    r = 2 * i + 2
    steps += 3 
    
    if l < n and arr[i] < arr[l]:
        largest = l
        steps += 1
   
    if r < n and arr[largest] < arr[r]:
        largest = r
        steps += 1
    
    if largest != i:
        arr[i],arr[largest] = arr[largest],arr[i]  
        steps += heapify_steps(arr, n, largest, steps) + 1 
    return(steps)

def heapSort_steps(arr):
    steps = 0
    n = len(arr)
  
    for i in range(n // 2 - 1, -1, -1):
        steps += heapify_steps(arr, n, i, steps) 
  
    for i in range(n-1, 0, -1):
        arr[i], arr[0] = arr[0], arr[i]
        steps += heapify_steps(arr, i, 0, steps)
        
    return(steps)
```
